### Encode the data for modeling

```python
# Encode the data for model training because machine can't understand categorical data
df['cut_num'] = df.cut.replace(['Fair','Good','Very Good','Premium','Ideal'],[1,2,3,4,5])
df
```

```python
df[10:50]
```

```python
# import necessary library for encoding data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Encode the data with LabelEncoder
df['color_num'] = le.fit_transform(df.color)
```

```python
df[10:50]
```

```python
# Encode the data for model training because machine can't understand categorical data
df['clarity_num'] = df.clarity.replace(['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF'],[0,1,2,3,4,5,6,7])
```

```python
df.head(10)
```

```python
# Devide DataFrame into x and y part
x_final = df.drop(['cut','color','clarity','price'],axis=1)
x_final
```

```python
y_final = df.price
y_final
```

### Data Split

```python
# split dataset into training and testing part
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final,y_final,test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

### Apply LinearRegression

```python
# import LinearRegression from sklearn and create object of it
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# train model using Linear Regression
lr.fit(x_train,y_train)
```

### Perfomane Check

```python
# check model accuracy using score method
lr.score(x_test,y_test)
```

```python
y_predicted = lr.predict(x_test)
```

```python
# import r2_score from sklearn for calculating R*R
from sklearn.metrics import r2_score
```

```python
# calculate R*R using r2_score
r2_score(y_test,y_predicted)
```

```python
# import mean_squared_error from sklearn
from sklearn.metrics import mean_squared_error
```

```python
# calculate mean squared error
mean_squared_error(y_test,y_predicted)
```
