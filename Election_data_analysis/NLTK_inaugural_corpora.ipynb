## Importing Libraries

```python
import nltk
nltk.download('inaugural')
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import inaugural
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

## Utilities

```python
def count_character_words_sentences(speech):
  text = inaugural.raw(speech)
  characters = len(text)
  words = inaugural.words(speech)
  words_count = len(words)
  sentences = inaugural.sents(speech)
  sent_count = len(sentences)
  print(f'{speech} contains:: Characters: {characters} :: Words: {words_count} :: sentences: {sent_count} ')
```

```python
def remove_stop_words(speech, stop_words):
  words = inaugural.words(speech)
  cleaned_words = [w for w in words if not w.lower() in stop_words]
  return cleaned_words
```

```python
def show_word_clound(words_count):
  wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(words_count)
  plt.figure(figsize=(10,6))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
  plt.close()
```

## 2.1 Find the number of characters, words, and sentences for the mentioned documents.

```python
count_character_words_sentences('1941-Roosevelt.txt')
count_character_words_sentences('1961-Kennedy.txt')
count_character_words_sentences('1973-Nixon.txt')
```

## 2.2 Remove all the stopwords from all three speeches. 

```python
stop_words = set(stopwords.words('english'))
```

```python
cleaned_1943 = remove_stop_words('1941-Roosevelt.txt', stop_words)
cleaned_1961 = remove_stop_words('1961-Kennedy.txt', stop_words)
cleaned_1973 = remove_stop_words('1973-Nixon.txt', stop_words)
```

## 2.3 Which word occurs the most number of times in his inaugural address for each president? Mention the top three words

```python
words_count_1943 = Counter(cleaned_1943)
print(words_count_1943)
```

```python
# President Franklin D. Roosevelt in 1941 Speech contains: know, spirit, life occurs the most.
```

```python
words_count_1961  = Counter(cleaned_1961)
print(words_count_1961)
```

```python
# President John F. Kennedy in 1961 speech contains: us, world, let occurs the most
```

```python
words_count_1973 = Counter(cleaned_1973)
print(words_count_1973)
```

```python
# President Richard Nixon in 1973 speech contains: us, America, peace occurs the most.
```

### 2.4 Plot the word cloud of each of the speeches of the variable. (after removing the stopwords)

```python
show_word_clound(words_count_1943)
show_word_clound(words_count_1961)
show_word_clound(words_count_1973)
```
